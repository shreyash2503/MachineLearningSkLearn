```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
```

```python
X.head()
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Training the model with default hyperparameters

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, base_preds))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, base_preds)
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Features Importances']).sort_values('Features Importances')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 10), dpi=400)
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize=(12, 10), dpi=400)
    plot_tree(model, feature_names=X.columns, filled=True);
    
```

### Training the model with combination of different hyperparameters

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train, y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train, y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train, y_train)
```

```python
report_model(entropy_tree)
```
